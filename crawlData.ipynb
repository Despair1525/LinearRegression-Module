```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
url = "https://batdongsan.com.vn/ban-nha-dat?gcn=30-ty&dtln=500m2"
driver.get(url)
print("Open Web succses")
```

```python
#list Link "re__srp-list js__srp-list"
page_source = BeautifulSoup(driver.page_source)
print(page_source)
```

```python
def getURL():
    page_source = BeautifulSoup(driver.page_source)
    listLink = page_source.findAll('a',class_="js__product-link-for-product-id" )
    allLinkList = []
    for link in listLink:
        urlID = link.get("href")
        linkURL = str("https://batdongsan.com.vn/")+ str(urlID)
        if linkURL not in allLinkList:
            allLinkList.append(linkURL)
    return allLinkList
```

```python
allLink=[]
for page in range(2500):
    url_one_page = getURL()
    allLink.append(url_one_page)
    script='window.scrollTo(0,200*'+str(len(url_one_page)+2)+")"
    driver.execute_script(script)
    next_button =driver.find_elements_by_class_name("re__icon-chevron-right--sm")[0]
    print("Done page "+str(page+1))
    next_button.click()
print(len(allLink))
```

```python
def saveFile(name,data,store=False):
    textfile = open(name+".txt", "w")
    list=[]
    for page in allLink:
        for link in page:
            print(link)
            list.append(link)
            textfile.write(link + "\n")
    textfile.close()
    if store:
        return list
listWebLink=saveFile("dataList",allLink,store=True)
```

```python
import re
def extract(soup):
    code='>(.+):<'
    resultr = re.findall(code,soup)
    if len(resultr)==0:
        resultr=re.findall('>(.+)<',soup)
    return resultr[0]
def extracLine(text):
    text=text.split("\n")
    return text
def getDICT():
    try:
        source = BeautifulSoup(driver.page_source)
    except:
        print("cant open source")
        return

    infoY = source.find('div',class_='re__pr-short-info js__pr-short-info').extract()
    text=extracLine(str(infoY.text))
    DateINFO =  source.find('div',class_='re__pr-short-info-item js__pr-config-item').extract()
    Date = DateINFO.text
    try:
        match = re.search(r'(\d+/\d+/\d+)',Date)
        price = text[text.index('Mức giá')+2]
        S = text[text.index('Diện tích')+1]
        Date=match.group(1)
    except:
        print("Not Found Price")
        return

    ########################################################################################
    info= source.findAll('div',class_='re__section-body re__border--std js__section-body')
    info = BeautifulSoup(str(info[0]))
    infoDetai = info.findAll('div',class_="re__list-standard-1line--md")
    dct={}
    if len(infoDetai) <=5 :
        return
    for i in infoDetai:
        soup = BeautifulSoup(str(i))
        title =extract(str(soup.find(class_="title")))
        value = extract(str(soup.find(class_="value")))
        title = title.lower()
        dct[title]=[value]
        dct["price"] =price
        dct["m^2"] = S
        dct["Date"] = Date
    return dct
```

```python
df=pd.DataFrame()
NumSuc=0
NumFail=0

def printInfo(NumSuc,NumFail):
    print("Total Web Load:("+str(NumFail+NumSuc)+")",end=" ")
    print("Succsess:",NumSuc,end="|")
    print("Fail: ",NumFail)
    print("")
count=0
for i in listWebLink:
    count +=1
    nextURL=i
    driver.get(nextURL)
    dct = getDICT()
    if dct is None:
        NumFail +=1
        printInfo(NumSuc,NumFail)
        continue
    else:
        df2=pd.DataFrame.from_dict(dct)
        df=df.append(df2,ignore_index=True)
    NumSuc +=1
    printInfo(NumSuc,NumFail)

```
